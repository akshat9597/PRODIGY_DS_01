```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots inside notebook
%matplotlib inline
```

```python
np.random.seed(42)

data = pd.DataFrame({
    "Age": np.random.randint(18, 60, 200),
    "Gender": np.random.choice(["Male", "Female"], 200),
    "Salary": np.random.randint(20000, 100000, 200),
    "Department": np.random.choice(
        ["IT", "HR", "Sales", "Finance"], 200)
})

data.head()
```

# New Section

```python
print(data.info())

print("\nSummary Statistics:")
print(data.describe())
```

```python
plt.figure()

plt.hist(data["Age"])

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure()

data["Gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure()

data["Department"].value_counts().plot(kind="pie")

plt.title("Department Distribution")

plt.ylabel("")

plt.show()
```

```python
plt.figure()

sns.heatmap(data.corr(numeric_only=True),
            annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure()

sns.histplot(data["Salary"])

plt.title("Salary Distribution")

plt.show()
```

Key Insights:

• Most employees are aged between 25–40
• Gender distribution is nearly balanced
• IT department has highest count
• Salary distribution shows moderate spread
• Age and Salary show weak correlation
